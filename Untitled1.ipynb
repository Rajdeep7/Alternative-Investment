```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lse = pd.read_csv('LSE.csv')
lse.plot(x='Year',y='Ending Price (in million GBP)', title = 'LSE')
```

```python
wine = pd.read_csv('Wine.csv')
wine = wine.fillna(wine['Price (in million GBP)'].mean())
wine.plot(x='Year', label = 'Wine', title = 'Wine')
```
